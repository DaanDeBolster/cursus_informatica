## 1. For</h2>


In dit hoofdstuk wordt na de `while`-lus ook de `for`-lus geïntroduceert, een alternatieve manier om controle in een Pythonprogramma te implementeren.

Iteratie betekent "overlopen". Er wordt dus een stuk code meerdere keren uitgevoerd. Bij elke keer dat de code wordt doorlopen (dus bij elke **Iteratie**) verandert de waarde van een bepaalde variabele. Welke waarden dit zijn wordt bepaald door een itereerbaar object.

> De syntax van een `for`-lus in Python is de volgende:
```python
for iterarievariabele in itereerbaar_object:
  code
```
> - Het **itereerbaar object** is het object waarvan de elementen worden overlopen. Elk object dat kan worden geïndexeerd is itereerbaar. Voorbeelden die we al kennen zijn:
  - Een lijst
  - Een tuple
  - Een string
- De **iteratievariabele** nemen per iteratie de waarde aan van het volgende element uit het itereerbaar object.
- Net als bij `if` en `while` wordt een indentatie gebruikt om aan te duiden welke code meerdere keren moet worden uitgevoerd. In zo goed als elk geval wordt in deze code de iteratievariabele opgeroepen.


#### Voorbeelden

Onderstaand voorbeeld print alle elementen uit een lijst op de console.

````python
lijst = [12, 17, 19.5, 18, 0.2, -7.1]
for element in lijst:
  print(element)
````

We kunnen op een analoge manier alle elementen tekens van een string op het scherm printen.

````python
tekst = "Dit is een tekst!"
for teken in tekst:
  print(teken)
````

#### Opmerking

In Python zijn selectie en iteratie equivalent. Dat wil zeggen dat elke code met een `while`-lus in code met een `for`-lus kan worden omgezet en omgekeerd. Uiteraard is één van de twee methoden vaak meer intuïtief dan de andere afhankelijk van het probleem.

Hieronder zijn de voorbeelden van hierboven herschreven met een `while`-lus.

````python
lijst = [12, 17, 19.5, 18, 0.2, -7.1]
n = 0
while n < len(lijst):
  print(lijst[n])
  n += 1
````

````python
tekst = "Dit is een tekst!"
n = 0
while n < len(tekst):
  print(tekst[n])
  n += 1
````

#### Voorbeeld

Hieronder nog een extra voorbeeld dat alle elementen uit een lijst kwadrateert.

````python
invoer = [2, 3, 5, -9, 13]
uitvoer = []
for i in invoer:
  kwadraat = i**2
  uitvoer.append(kwadraat)
  print(kwadraat)
  print(uitvoer)
  print("")
print(uitvoer)
````

## 2. Range

Vaak wil je over een regelmatige reeks getallen itereren. Je hoeft hiervoor niet steeds een nieuwe lijst te constueren. Een alternatief is het itereerbaar object `range` gebruiken.

> Een **range-object** is een opeenvolgende reeks getallen:
- `range(n)` levert alle gehele getallen van `0` tot `n-1`. (Er worden dus `n` getallen gegenereert)
- `range(m,n)` levert alle gehele getallen van `m` tot `n-1`.  (Er worden dus `n-m` getallen gegenereert)

#### Voorbeeld

Onderstaande code print van twee lijsten met gelijke lengte een nieuwe lijst op het scherm die per element het maximum neemt van dezelfde elementen uit de twee ingevoerde lijsten.

````python
lijst1 = [2, 9, -2, 18, 90]
lijst2 = [7, 2, 12, 99, 90]

lijst_max = []
for i in range(len(lijst1)):
  maximum = max(lijst1[i], lijst2[i])
  lijst_max.append(maximum)
print(lijst_max)
````

#### Oefening 2

## 3. List comprehention

Een handige manier om snel lijsten te bouwen met compacte code is via **list comprehention**. Dit wil zeggen dat onderstaande code...
```python
lijst = []
for element in itereerbaar_object:
  lijst.append(bewerkt_element)
```
... equivalent is aan het onderstaande.
```python
lijst = [bewerkt_element for element in itereerbaar_object]
```

#### Voorbeeld

Onderstaande code genereert een lijst van de kwadraten van 1 tot 10 via list comprehention.

````python
kwadraten = [n**2 for n in range(1,11)]
print(kwadraten)
````

## 4. Geneste lussen

Geneste lussen zijn lussen die binnen een andere lus worden uitgevoerd. Nesting van lussen kan een probleem snel een pak complexer maken. Hou dus zeker goed bij hoeveel keer elke lus wordt uitgevoerd.

#### Voorbeeld

Onderstaande code print de maaltafels van 1 tot 10 op de console via 2 geneste for-lussen.

````python
for i in range(1,11):
  tafel = "De tafel van " + str(i) + " is:\t"
  for j in range(1,11):
    tafel += str(i*j)
    tafel += " "
  print(tafel)
````

Je kan ook while-lussen nesten, of for- en while-lussen samen. Onderstaande code is equivalent.

````python
for i in range(1,11):
  tafel = "De tafel van " + str(i) + " is:\t"
  j = 1
  while j <= 10:
    tafel += str(i*j)
    tafel += " "
    j += 1
  print(tafel)
````

## 5. Insertionsort

Een zeer divers probleem binnen de algoritmiek is het sorteerprobleem.

> Bij het **sorteerprobleem** moet een eindige lijst van reëele getallen worden gesorteerd van klein naar groot.

Met andere woorden: We zoeken een algoritme met als invoer een willekeurige lijst reële getallen, en als uitvoer een lijst met dezelfde getallen, maar gesorteerd, wetende dat we 2 getallen met elkaar kunnen vergelijken via `>`, `<`, `=`,...

Dit probleem wordt vaak gebruikt om algoritmische technieken (zoals hier iteratie) te demonstreren. Er bestaan dan ook zeer veel algoritmen om dit probleem op te lossen, elk met hun voor- en nadelen.

> Het meest intuitieve algoritme om het sorteerprobleem op te lossen is **insertionsort**.

Insertionsort werkt op dezelfde manier zoals je bijvoorbeeld een aantal speelkaarten in je hand zou sorteren. Je overloopt alle elementen in een lijst van links naar rechts. Elk element verplaats je dan naar links tot alle elementen links van de originele plaats van het element zijn gesorteerd.

**Visualisatie**

**Code**

In onderstaande code is insertionsort geïmplementeerd. Voer ze zeker eens uit voor verschillende inputs en bestudeer alle tussenstappen.

````python
lijst = [12, 13, 17, 11, 16]                    # Voer een lijst in

# Insertionsort
for i in range(len(lijst)):                     # Overloop element per element het ongesorteerd gedeelte
    print(" ")
    waarde = lijst[i]                           # Neem element i apart
    for j in range(0, i):                       # Vergelijk het element met de elementen in het gesorteerde deel
        if waarde < lijst[j]:                   # Heb je de positie van het element gevonden? ...
            waarde, lijst[j] = lijst[j], waarde # ... Verwissel die en loop verder met een nieuw element
        print(lijst)                            # Print per stap het tussenresultaat op het scherm (niet noodzakelijk, wel handig om het algortitme te begrijpen)
    lijst[i] = waarde


print("\nGesorteerde lijst:", lijst)
````

#### Eigenschappen

Om een lijst met `n` elementen te sorteren via insertionsort zijn er `n*(n-1)/2` iteraties nodig. Het is duidelijk dat het aantal iteraties kwadratisch stijgt in functie van de lengte van de te sorteren lijst. Hoewel insertionsort een van de meest eenvoudige sorteeralgoritmen is, is het zeker niet het snelste. In de praktijk wordt het vaak gebruikt als de te sorteren lijst kort of al deels gesorteerd is.

Insertionsort is een goed voorbeeld van een **incrementeel algoritme**. Dat wil zeggen dat elke iteratie al een geldige tussenoplossing levert. In dit geval wil dat zeggen dat, op gelijk moment in het algoritme, alle elementen links van het geselecteerde element gesorteerd zijn.
