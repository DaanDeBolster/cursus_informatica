### 2.3 if-else

Aan een `if`-statement kan telkens een `else`-statement worden toegevoegd om alternatieve code te specifiëren die wordt uitgevoerd als de conditie vals is.

> ```python

### 2.5 if-elif-else

Je kan meerdere alternatieve condities aan elkaar koppelen via het `elif`-sleutelwoord (else if).

> ```python
if conditie1:
  code1
elif conditie2:
  code2
else:
  code3
```
> Eerst wordt `conditie1` geëvalueerd. bij `True` wordt `code1` uitgevoerd. Bij `False` wordt `conditie2` geëvalueerd. Als die conditie `True` oplevert wordt `code2` uitgevoerd, anders wordt `code3` uitgevoerd.

ITE.drawio (1).png

Merk op dat de bovenstaande code equivalent is aan

```python
if conditie1:
  code1
else:
  if conditie2:
    code2
  else:
    code3
```

Je kan ook meerdere dan 2 voorwaarden aan je code toevoegen via `elif`:

```python
if conditie1:
  code1
elif conditie2:
  code2
elif conditie3:
  code3
#...
else:
  codeN
```


#### Voorbeeld

Onderstaand voorbeeld herschrijft de code uit het voorbeeld in de vorige sectie met `elif`. Dit levert code op die een stuk leesbaarder is.

````python
T = float(input("Geef een temperatuur in °C: "))
if T <= 0:
  print("Het water is vast.")
elif T >= 100:
  print("Het water is een gas.")
else:
  print("Het water is vloeibaar.")
````

## 3. Gevallenanalyse

Een eerste eenvoudige, echter zeer inefficiënte, algoritmische techniek is gevallenanalyse.

> Bij **gevallenanalyse** wordt elke mogelijke oplossing van een bepaald probleem overlopen en gecontroleerd of de oplossing correct is. <br><br>
> Een algoritme via gevallenanalyse vereist dus twee "ingrediënten":
- Een lijst van alle mogelijke oplossingen.
- Een test om te controleren of een oplossing correct is.

#### Voorbeeld

Onderstaand algoritme berekend het maximum van drie getallen via gevallenanalyse.

````python
getal1 = float(input("Geef een eerste getal: "))
getal2 = float(input("Geef een tweede getal: "))
getal3 = float(input("Geef een derde getal: "))

if getal1 > getal2 and getal1 > getal3:     # 1e geval: getal1 is het maximum
  print(getal1)
elif getal2 > getal1 and getal2 > getal3:   # 2e geval: getal2 is het maximum
  print(getal2)
else:                                       # 3e geval: getal3 is het maximum
  print(getal3)
````

Merk op dat dit algoritme (voorlopig) enkel voor een klein aantal getallen haalbaar is. Denk maar aan het programmeerwerk bij bijvoorbeeld 4, 5 of n getallen.

Merk ook op dat dit algoritme, net als vele andere algoritmen, al in Python zit ingebouwd via de `max()` functie. Onderstaande code is dus equivalent (en een stuk leesbaarder).

````python
getal1 = float(input("Geef een eerste getal: "))
getal2 = float(input("Geef een tweede getal: "))
getal3 = float(input("Geef een derde getal: "))

maximum = max(getal1, getal2, getal3)
print(maximum)
````
