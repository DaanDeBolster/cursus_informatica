## 3. Variabelen

> Een **variabele** is een tijdelijke plaats in het geheugen waarin een object (bijvoorbeeld een getal of tekst) kan worden bijgehouden.
> Wanneer we werken met variabelen beschouwen we twee operaties: Een **toekenning** en een **oproep**.
- Een nieuwe variabele wordt automatisch gecreërd wanneer we de eerste keer object toekennen. We gebuiken hierbij de `=`-operator. Je geeft de variabele dan een duidelijke en zinvolle naam.
- Door de naam van de variabele in je code te verwerken kan je het object dat in de variabele zit opgeslagen oproepen.

#### Voorbeelden

In onderstaand voorbeeld creëren we een variabele met als naam `getal` en kennen we als object de waarde `8` toe.

````python
getal = 8
````

9iR5LrypT.png

Je kan dan verderop in je code de variabele oproepen. Je gebruikt dan de waarde die in de variabele is opgeslagen.

````python
getal = 8
print("Het getal x is gelijk aan", getal)
````

Je kan de waarde die in een variabele zit ook gebruiken in de toekenning van een andere variabele.

````python
getal = 8
getal_plus_twee = getal + 2
print("Twee meer dan", getal, "is", getal_plus_twee)
````

Je kan ook het resultaat van een operatie op een variabele terug in dezelfde variabele opslaan.

````python
getal = 8
getal = getal + 2
print("Twee meer is", getal)
````

Dit kan echter nog eenvoudiger met samengestelde operatoren.

> Via **samengestelde opereratoren** kan je voor een variabele een operatie op zichzelf definiëren: `+=`, `-=`, `*=`, `/=`,... <br>
Zo is bijvoorbeeld `x -= 3` equivalent aan `x = x - 3`.

#### Voorbeeld

````python
getal = 7
getal += 2
print("Twee meer is", getal)
````

## 4. Datatypes

In de vorige paragraaf werd een variabele beschereven als een stuk geheugen dat een stuk data in de vorm van een object bijhoud. Zo'n object kan verschillende vormen aannemen.

> Een **datatype** beschrijft wat voor soort informatie een object voorstelt.

**[ℹ]** In Python zijn 15 verschillende datatypes gedefiniëerd:
`str, int, float, complex, list, tuple, range,	dict, set, frozenset, bool, bytes, bytearray, memoryview, NoneType`

> We behandelen hier drie soorten datatypes:
- `str` = Een **string** = Een reeks symbolen die een text kunnen vormen
  - Bijvoorbeeld: `"Dit is een tekst"`, `"Hello world!"`, `"!"`,...
- `int` = Een **integer** = Een geheel getal
  - Bijvoorbeeld: `"1000000"`, `"12"`, `"-396"`,...
- `float` = Een **floating point number** = Een reëel getal
  - In Python worden kommagetallen met een `.`aangeduid.
    - Bijvoorbeeld: `-14.52`, `.1263`, `-200.0`,...
  - Onderstaande voorbeelden duiden de wetenschappelijke notatie aan.
    - Bijvoorbeeld: `-21.13e12` = -21,13·10¹²

Basic-Data-Types-in-Python_Watermarked copy.png

> Via de functie `type()` kan je het datatype van een object checken.

#### Voorbeeld

````python
a = "Dit is een tekst"
b = -75
c = 12.9

print(type(a))
print(type(b))
print(type(c))
````

**[⚛]** In andere programmeertalen moet vaak bij het aanmaken van een variabele een datatype worden meegegeven. In Python gebeurt dit automatisch. De interpreter kiest zelf het datatype van een object aan de hand van zijn kenmerken. Dit concept wordt ook wel **duck typing** genoemd.

````python
a = -94
b = -94.

print(type(a))
print(type(b))
````

**[⚛]** Bovendien kan het datatype van een object veranderen. Dit heet **dynamic typing**.

#### Voorbeeld

Door een `int` met een `1.2` te vermenigvuldigen krijgen we een `float`.

````python
a = 10
print(type(a))

a *= 1.2
print(type(a))
````

## 5. Modules

Uiteraard kan maar een beperkte functionaliteit aan een programmeertaal worden toegevoegd. Merk bijvoorbeeld op dat in Python een bewerking ontbreekt om de worteltrekking uit te voeren. Hiervoor doet men beroep op modules.

> Een **module** is een bibliotheek van extra functionaliteit die aan een programma kan worden toegevoegd. Je moet vooraf aangeven wanneer je een bepaalde module aan je programma wil toevoegen. Dit kan met het `import` sleutelwoord.

#### Voorbeeld

Om de worteltrekking in Python toe te passen importeren we de module `math`. Deze module bevat allerlei extra wiskundige bewerkingen: https://docs.python.org/3/library/math.html

In de documentatie lezen we dat we de worteltrekking kunnen uitvoeren via de `sqrt()` functie.

````python
import math       # Module importeren

x = 49
wortel = math.sqrt(x)
print("De vierkantswortel van", x, "is", wortel)
````

Merk op dat we steeds het de naam van de module als voorvoegsel aan de functie `sqrt` moeten toevoegen om te specifiëren dat deze uit de `math` module komt.

We kunnen dit ook vergemakkelijken door een nieuwe (kortere) naam voor de module door te voeren.

````python
import math as m      # Module importeren en een nieuwe naam geven

x = 49
wortel = m.sqrt(x)
print("De vierkantswortel van", x, "is", wortel)
````

Als we, zoals in dit geval, maar één functie uit de module gebruiken hoeven we ook niet heel de module in te laden. Je kan bij het importeren aangeven welke functies je nodig hebt.

````python
from math import sqrt       # Importeer enkel sqrt uit de module math

x = 49
wortel = sqrt(x)      # De module nog eens specifieren is nu niet langer nodig
print("De vierkantswortel van", x, "is", wortel)
````
