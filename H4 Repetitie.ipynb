## 2. While

Herhaling kan in Python worden geïmplementeerd via de `while`-lus.
> ```python
while conditie:
  code
```
Hier wordt `code` uitgevoerd zolang `conditie` `True` oplevert. Telkens `code` is uitgevoerd evalueert het programma `conditie` opnieuw. Als de conditie `False` oplevert springt het programma uit de lus.

#### Voorbeeld

Onderstaande code halveert een getal tot het kleiner is dan 1 en print het resultaat op het scherm.

`````python
getal = float(input("Geef een getal: "))
while getal >= 1:
  print(getal)
  getal /= 2

print(getal)
`````

### 2.1 Oneindige lus

> Een **oneindige lus** is een lus zonder stopconditie. In het geval van `while` dit zeggen dat de conditie altijd `true` oplevert. In de logica noemen we dit een **tautologie**.



Typisch is een oneindige lus iets wat we in een programma willen vermijden. Controleer dus steeds of je `while`-lus een stopconditie bevat.

#### Voorbeeld

`````python
print("Daft Punk - Around the world (lyrics):\n")
while True:
  print("Around the world")
`````

### 2.2 Telvariabele

> Een **telvariabele** is een variabele waarvan de waarde elke keer dat de lus wordt doorlopen verandert.</br></br>
Een startwaarde aan de telvariabele toekennen noemen we **initialisatie**.
Bij de telvariabele een waarde optellen noemen we **incrementatie**. Typisch start een telvariabele op 0 en wordt elke keer dat de lus wordt doorlopen 1 bij de telvariabele opgeteld.

#### Voorbeeld

Onderstaande code print de maaltafel van een geheel getal op het scherm.

`````python
p = int(input("Geef de tafel van "))
n = 0             # Initialiseren van de telvariabele op 0
while n <= 10:
  print(p, "maal", n, "is", p * n)
  n += 1          # Incrementeren van de telvariabele met 1
`````

## 3. Algoritme van Euclides

> Het **Algoritme van Euclides** is één van de bekendste algoritmen binnen de discrete wiskunde. Het neemt twee gehele getallen als invoer en geeft de grootst gemene deler van de twee getallen terug.

Zoals bij elk algoritme kan ook het Algortitme van Euclides via verschillende voorstellingen worden beschreven:

Stappenplan:
1. Noem beide gehele getallen *a* en *b*.
2. Trek van het grootste van de twee getallen het kleinste af.
3. Herhaal de tweede stap tot beide getallen gelijk zijn.
4. De resulterende (gelijke) waarde van beide getallen is de grootst gemeenschappelijke deler.

Python code:

`````python
a = int(input("Geef een 1e getal: "))
b = int(input("Geef een 2e getal: "))

while a != b:
  if a > b:
    a -= b
  else:
    b -= a

print("De ggd is", a)
`````
