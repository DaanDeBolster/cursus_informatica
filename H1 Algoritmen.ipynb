## 1. Het algoritme



> Een **algoritme** is een omschrijving van een manier waarop een probleem kan worden opgelost of een taak kan worden uitgevoerd. Typisch kan een algoritme worden opgedeeld in een reeks instructies of regels die in een vaste volgorde moet worden uitgevoerd.

Hoewel het begrip *algoritme* zeer vaak wordt gebruikt in de context van de informatica, hoeft dit niet noodzakelijk het geval te zijn. Wanneer we een algoritme zien als een reeks instructies, of een stappenplan, is de link met het dagelijkse leven niet ver te zoeken.

## 1.1 Beschrijven van een algoritme

Uit de voorbeelden hierboven is duidelijk dat er verschillende manieren zijn om een algoritme te beschrijven. Hiervoor wordt er zeer vaak voor één van de onderstaande opties gekozen.

> * Een **tekstuele omschrijving** maakt via een platte tekst de instructies uit een algoritme duidelijk.
* Een **stroomdiagram** of flowchart is een schematische weergave van een algoritme.
* **Computercode**, of meestal gewoon *code*, beschrijft een algoritme via een programmeertaal, die een computer kan begrijpen. Daarover later meer.
* **Pseudocode** is een alternatief voor computercode, dat niet door een computer kan worden uitgevoerd.

#### Voorbeeld

In dit voorbeeld wordt een bepaald algoritme in de vier bovenstaande voorstellingen beschreven.

***Tekstuele omschrijving***

Het algoritme bepaalt aan de hand van de temperatuur of het een koude, normale of warme zomerdag is in Amerika. Dat gebeurt in verschillende stappen:
- Geef de temperatuur in graden Fahrenheit.
- Bereken de temperatuur in graden Celsius = (Fahrenheit - 32) ÷ 1.8
- Is de temperatuur lager dan 20 dan is het koud. Is de temperatuur hoger dan 30 dan is het warm.

***Stroomdiagram***

Screenshot 2024-11-26 at 09.47.16.png

***Computercode***

Onderstaand codeblok kan worden uitgevoerd met `shift+enter`.

````python
temp_F = int(input("Geef een temperatuur in graden Farenheit: "))
temp_C = (temp_F-32)/1.8
if temp_C < 20:
  print("Koude dag")
elif temp_C > 30:
  print("Warme dag")
else:
  print("Normaal")
````

***Pseudocode***

Het nadeel van code is dat je aan de syntax (de "taalregels") van een programmeertaal bent verbonden. Wat het algoritme vaak moeilijk leesbaar maakt. Als we een algoritme willen beschrijven met code maar deze niet per-se met een computer willen uitvoeren wordt dus vaak voor pseudocode gekozen.




```
F = Invoer(Geef een temperatuur in graden Farenheit.)
C = (F-32)/1.8
Als C<20:
	Uitvoer(Koude dag)
Als C>30:
	Uitvoer(Warme dag)
Anders:
	Uitvoer(Normaal)
```

## 2. Het programma</h2>


### 2.1 Algoritme VS Programma

Het is belangrijk om het verschil tussen een algoritme en een programma in te zien:
- Een algoritme is reeks instructies om een probleem op te lossen
- Een programma is een vertaling van een algoritme in een taal die een computer kan verstaan, of een programmeertaal.

### 2.2 Programmeertalen

> Een **programmeertaal** is taal waarin een gebruiker kan vastleggen hoe een computer een algortitme moet uitvoeren

Een programmeertaal  is ontworpen om makkelijk door mensen te worden gebruikt. Ze is dus ook vaak opgebouwd met intuïtieve sleutelwoorden zoals (`if, else, for, while, break, jump, exit`,...).

Een computer verstaat uiteraard geen Engels. Een programma in een programmeertaal moet dus vertaald worden naar een andere reeks instructies die door de processor van de computer kan worden verstaan en uitgevoerd. Die reeks instructies noemen we **machinetaal**. Het programma dat code in een programmeertaal vertaald naar machinetaal noemen we een **compiler**. Het proces van code naar machinetaal vertalen noemen we **compileren**.
