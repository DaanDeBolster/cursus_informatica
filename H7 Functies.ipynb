## 1. Functies

#### Oefening 1

Schrijf een programma dat de oppervlakte van twee cirkels vergelijkt. De invoer zijn de stralen van de twee cirkels. De uitvoer is het verschil in oppervlakte tussen de twee cirkels.

Met de kennis die we op dit moment hebben kunnen we niet anders dan de formule voor de oppervlakte van een cirkel twee keer te implementeren. Om zo'n herhaling te vermeiden kan een functie worden gedefiniëerd.

> Een **functie** is een afzonderlijke reeks instructies die meerdere keren kan worden opgeroepen.

Voordat een functie kan worden opgeroepen moet ze worden gedefiniëerd. Met andere woorden: We schrijven afzonderlijk een stuk code dat beschrijft wat een functie precies moet doen om een bepaalde invoer in de juiste uitvoer om te zetten. Een functie wordt niet voor niks ook vaak een *subprogramma* of *subroutine* genoemd.

> Een **functiedefinitie** bakent de procedures af die later kunnen worden opgeroepen. In Python gebruikt men hiervoor de `def`-structuur als volgt:
```python
def functienaam(argumenten):
  procedure
  return resultaat
```
- De **functienaam** is een naam die je aan je functie geeft opdat je ze later via dezelfde naam opnieuw kan oproepen.
- De **argumenten** zijn stellen de invoer van de functie voor als een reeks variabelen die bij de functieoproep worden geïnitialiseerd.
- De **procedure** is een reeks instructies die bewerkingen op de argumenten toepassen.
- Via een **return**-statement kan eventueel een uitvoerwaarde worden teruggegeven. Het doorlopen van de functieprocedure wordt steeds onderbroken bij na het uitvoeren van *return*.


#### Voorbeeld

In de onderstaande code wordt een functie `meest_dichtbij` gedefiniëerd met drie argumenten `a`, `b` en `x`. De functie geeft terug welke van de twee getallen `a` en `b` het dichtste bij een derde getal `x` ligt.

````python
def meest_dichtbij(a, b, x):
  if abs(a-x) < abs(b-x):
    dichtste = a
  else:
    dichtste = b
  return dichtste
````

> Nadat je een functie hebt gedefiniëerd kan je ze oproepen via een **functieoproep**. Dat kan als volgt:
```python
uitvoer = functienaam(invoer)
```

#### Voorbeeld

Nu de functie `meest_dichtbij` is gedefiniëerd kan ze worden opgeroepen.

````python
test_dichtste1 = meest_dichtbij(2, 9, 10)
test_dichtste2 = meest_dichtbij(100, -100, -0.5)

print(test_dichtste1)
print(test_dichtste2)
````

## 2. Standaardargumenten

Standaardargumenten worden achteraan de argumentenlijst toegevoegd en krijgen daar een standaardwaarde toegekent via de `=`-operator. Bij een functieoproep  hoeft het standaardargument dan niet worden meegegeven. Indien dat het geval is wordt in de procedure de standaardwaarde gebruikt. Op deze manier kan je functies definiëren met een variabel aantal argumenten.

#### Voorbeeld

Onderstaande functie `rekening` telt de prijzen van een hoop artikelen aan de kassa op om een totale kostprijs te bekomen. Hiervoor neemt de functie een lijst `artikelen` als argument. Optioneel kan als tweede argument `korting` een kortingspercentage worden meegegeven. Indien dit argument niet wordt meegegeven wordt er geen korting toegepast.

````python
def rekening(artikelen, korting = 0):
  factor = 1 - korting/100
  return sum(artikelen)*factor

artikelen_1 = [12.99, 0.89, 2.09, 2.09]
artikelen_2 = [20.99, 3.99]

print(rekening(artikelen_1))      # Zonder korting (standaardargument `korting = 0` wordt gebruikt)
print(rekening(artikelen_1, 20))  # Met 20% korting

print(rekening(artikelen_2))      # Zonder korting (standaardargument `korting = 0` wordt gebruikt)
print(rekening(artikelen_2, 10))  # Met 10% korting
````

#### Voorbeeld

Bestudeer de errorboodschap die bij onderstaande codeblokken wordt gegenereerd.

````python
def f(x, y):
  z = x + y
  return z**2 - x
a = f(1, 2)
b = z
````

De compiler zegt dat `x` en `z` niet gedefiniëerd zijn. Toch worden zij geïnitialiseerd op respectievelijk de eerste en tweede lijn. De reden is dat de variabelen `x` en `z` na het uitvoeren van een functieprocedure worden vernietigd. Wanneer we dus later in de code (op de vijfde lijn) de waarde van variabele oproepen is er geen waarde meer beschikbaar. We zeggen dan dat de oproep buiten de *scope* van de variabele ligt.

> De **scope** of het **bereik** van een variabele is het deel van een programma waarin variabele kan worden opgeroepen.

De scope van `x` en `z` loop dus tot de derde lijn code. We noemen dit soort variabelen *lokale variabelen*.

> Een **lokale variabele** is een variabele waarvan de scope beperkt is binnen een functieprocedure. Een variabele is lokaal als...
- ...de variabele een argument van een functie is.
- ...de variabele binnen de procedure van een functie is geïnitialiseerd. <\br>

>Een **globale variabele** is een variabele die, vanaf zijn initialisatie in het gehele programma kan worden opgeroepen.

#### Voorbeeld

````python
def f(x, y):
  z = x + y
  return z**2 - b
b = 10
a = f(1, 2)
````

## 4. Functioneel programmeren

> **Functioneel programmeren** is een programmeerstijl waarbij een computerprogramma zo veel mogelijk uit functies is opgebouwd.

Het gebruik van van functies heeft een een hoop voordelen die hieronder staan beschreven.

### 4.1 Redundantie

Een goede programmeur hanteert tijdens het coderen altijd het DRY-principe.

> Het **DRY-principe** (Don't repeat yourself) houdt in dat in een computerprogramma zo weinig mogelijk code (idealiter zelfs geen code) meer dan 1 keer mag voorkomen.

Kopiëren en plakken van code maakt een computerprogramma langer en minder leesbaar. Bovendien wordt debuggen een pak moeilijker als je een foute lijn code gaat hergebruiken. Algemeen zeggen we dat het DRY-principe hanteren de redundantie van een programma verlaagt.

> **Redundante code** is code die voor de logica van een programma overbodig is.

Met functies kunnen we stukken code meer dan 1 keer oproepen en is kopiëren niet nodig.

Merk op dat ook de invoer van lussen de redundantie van code verlaagd: Code kan via repetitie en interatie meer dan 1 keer worden uitgevoerd.

### 4.2 Abstractie

Eens een functie is gedefiniëerd kan ze door andere gebruikers worden opgeroepen zonder dat zij de implementatie van de functie kennen. Zij hoeven enkel te weten wat de functie doet, niet hoe ze werkt.

Een goed voorbeeld is de `print` functie, die je ongetwijfeld al veel hebt gebruikt, maar waarvan je de werking wellicht nog steeds niet kent. Dit principe heet abstractie.

> **Abstractie** is de werking van een systeem verbergen om het gebruik ervan te vereenvoudigen.

Een programma met voldoende abstractie laat ons toe om stapsgewijze verfijning toe te passen.

> **Stapsgewijze verfijning** is een programmeerprincipe waarbij een probleem wordt voorgesteld als een samenstelling van meerdere eenvoudigere problemen.

#### Voorbeeld

Wanneer je wordt gevraagd om met je klas een fuif te organiseren lijkt dit op het eerste gezicht een complexe opdracht. Je maakt dus afspraken over welke leerling welk deel van de opdracht volbrengt. Dit maakt dan dat...

"Alle leerlingen uit de klas organiseren een fuif." </br>
⇕</br>
"Eén leerling regels DJ's" en</br>
"Eén leerling reserveert een zaal" en</br>
"Eén leerling maakt promo" en</br>
"Eén leerling besteld het drukwerk" en</br>
"Eén leerling besteld de drank" en</br>
"Eén leerling vraagt SABAM aan" en</br>
...

Via stapsgewijze verfijning is 1 complex probleem opgedeeld in meerdere werkbare problemen, die op hun beurt ook opnieuw kunnen worden verdeeld.

Goede voorbeelden van abstractie is hier:</br>
- De leerling die de promo doet moet weten welke DJ's op de fuif spelen, zodat hij de line-up op de affiche kan zetten. Maar hij hoeft hiervoor niet te weten hoe de andere leerling de DJ's heeft geregeld.
- De leerling die het drukwerk doet moet een ontwerp van de affiche ontvangen van de leerling die de promo doet. Maar hij hoeft niet te weten hoe deze leerling de affiche heeft ontworpen.
- ...
